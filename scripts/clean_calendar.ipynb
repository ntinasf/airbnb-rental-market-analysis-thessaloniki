```python
cal["available"] = cal["available"].map({'t': True, 'f': False})

type_dict = {
    "listing_id" : "object",
    "date" : "datetime64[ns]",
    "price" : "float32",
    "minimum_nights" : "Int16",
    "maximum_nights" : "Int16",

}

cal = cal.astype(type_dict)
```

```python
cal.drop(["price" ,"adjusted_price"], axis=1, inplace=True)
cal.info(memory_usage='deep')
```
